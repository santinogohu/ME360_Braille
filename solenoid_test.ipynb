```python
# solenoid_test_fan.py
import serial
import time

PORT = "COM3"
BAUD = 250000

ser = serial.Serial(PORT, BAUD, timeout=1)
time.sleep(2)  # allow board reset

def send(cmd):
    print(">>", cmd)
    ser.write((cmd + "\n").encode())
    ser.flush()
    while True:
        line = ser.readline().decode(errors="ignore").strip()
        if line:
            print("<<", line)
        if "ok" in line.lower():
            break

# optional: relax motion etc., but no axis moves here
send("G90")  # absolute, just to be safe

for i in range(1):
    print(f"Pulse {i+1}")
    send("M42 P65 S130")  # FAN ON -> IN1 HIGH -> solenoid ON
    time.sleep(0.2)
    #send("M42 P65 S0")       # FAN OFF -> IN1 LOW -> solenoid OFF
    #time.sleep(0.5)

ser.close()
print("Done.")
```
